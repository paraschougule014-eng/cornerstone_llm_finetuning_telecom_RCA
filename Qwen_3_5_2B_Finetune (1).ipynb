### Unsloth

````python
from unsloth import FastLanguageModel
import torch

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Qwen3.5-2B",
    max_seq_length = 4096, # Increased to handle longer thinking traces
    load_in_4bit = False,  # QLoRA not recommended for Qwen3.5; use 16-bit LoRA
    load_in_8bit = False, # [NEW!] A bit more accurate, uses 2x memory
    full_finetuning = False, # [NEW!] We have full finetuning now!
    dtype = torch.bfloat16

)
````

We now add LoRA adapters so we only need to update a small amount of parameters!

````python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Increased from 8 for better accuracy on 2B model
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 16,
    lora_dropout = 0, # Supports any, but = 0 is optimized
    bias = "none",    # Supports any, but = "none" is optimized
    # [NEW] "unsloth" uses 30% less VRAM, fits 2x larger batch sizes!
    use_gradient_checkpointing = "unsloth", # True or "unsloth" for very long context
    random_state = 3407,
    use_rslora = False,  # We support rank stabilized LoRA
    loftq_config = None, # And LoftQ
)
````

<a name="Data"></a>
### Data Prep
We use a custom `lightning_output.csv` dataset with three key columns:
- **question**: the user prompt
- **thinking**: the chain-of-thought reasoning (placed inside `<think>` tags)
- **reasoning**: the final answer

The assistant response is formatted as `<think>\n{thinking}\n</think>\n\n{reasoning}` so the model learns to produce thinking traces at inference time. Qwen 3.5 renders multi-turn conversations like below:


```
<|im_start|>user
Hello!<|im_end|>
<|im_start|>assistant
<think>\nLet me think...\n</think>\n\nHey there!<|im_end|>

```

First, upload `lightning_output.csv` to Colab (run the cell below), or mount Google Drive and adjust the path.

````python
# ---------- DATASET LOADING ----------
# Upload lightning_output.csv using the Colab file picker
# Alternatively, mount Google Drive and set csv_path to the file's Drive path
# from google.colab import files as colab_files
# uploaded = colab_files.upload()  # select lightning_output.csv when prompted

import pandas as pd
from datasets import Dataset

csv_path = "lightning_output_combined.csv"  # update if uploaded to a different path
df = pd.read_csv(csv_path, usecols=["question", "thinking", "reasoning"])
df = df.dropna(subset=["question", "thinking", "reasoning"]).reset_index(drop=True)
print(f"Loaded {len(df)} rows after dropping nulls")
dataset = Dataset.from_pandas(df)
````

We convert the dataset into conversational format. The `thinking` column is wrapped in `<think>...</think>` tags and the `reasoning` column becomes the final answer. Training on this format teaches the model to think before answering:

````python
def generate_conversation(examples):
    questions  = examples["question"]
    thinkings  = examples["thinking"]
    reasonings = examples["reasoning"]
    conversations = []
    for question, thinking, reasoning in zip(questions, thinkings, reasonings):
        # Wrap thinking in <think> tags; reasoning is the final answer
        assistant_content = f"<think>\n{thinking}\n</think>\n\n{reasoning}"
        conversations.append([
            {"role" : "user",      "content" : question},
            {"role" : "assistant", "content" : assistant_content},
        ])
    return { "conversations": conversations, }

dataset = dataset.map(generate_conversation, batched = True)
````

We now have to apply the chat template for `Qwen 3.5` onto the conversations, and save it to `text`.

````python
def formatting_prompts_func(examples):
   convos = examples["conversations"]
   texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False) for convo in convos]
   return { "text" : texts, }

dataset = dataset.map(formatting_prompts_func, batched = True)
````

<a name="Train"></a>
### Train the model
Now let's train our model on the custom `lightning_output.csv` dataset for 1 full epoch (~2283 rows). To do a quick test run first, uncomment `max_steps = 60` and comment out `num_train_epochs`.

````python
# Training on 2B model with custom dataset (~2283 rows, 1 epoch)
from trl import SFTTrainer, SFTConfig
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    eval_dataset = None, # Can set up evaluation!
    args = SFTConfig(
        dataset_text_field = "text",
        dataset_num_proc = 4, # Increasing "might" throw error on Colab/other envs.
        per_device_train_batch_size = 8,  # Reduced from 4 for Colab GPU compatibility
        gradient_accumulation_steps = 2,  # Effective batch size = 2 x 4 = 8
        warmup_steps = 5,
        num_train_epochs = 3, # Full training run over the dataset
        # max_steps = 60,  # Uncomment to do a quick test run instead
        learning_rate = 2e-4, # Reduce to 2e-5 for longer training runs
        logging_steps = 10,
        optim = "adamw_torch",
        weight_decay = 0.001,
        lr_scheduler_type = "linear",
        seed = 3407,
        report_to = "none", # Use TrackIO/WandB etc,
        bf16 = True,                    # H100 has native BF16 support; faster & more stable than fp16
        tf32 = True,                    # Enables TF32 for matmuls on Ampere+; free throughput gain
        gradient_checkpointing = True, # You have the VRAM — disable this for faster training
        dataloader_pin_memory = True,   # Faster CPU→GPU transfers
        dataloader_num_workers = 4,
    ),
)
````

We also use Unsloth's `train_on_completions` method to only train on the assistant outputs and ignore the loss on the user's inputs. This helps increase accuracy of finetunes!

````python
from unsloth.chat_templates import train_on_responses_only
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<|im_start|>user\n",
    response_part = "<|im_start|>assistant\n<think>",
)
````
